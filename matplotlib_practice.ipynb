```python
# import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "March", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars acclumated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create a plot using the MATLAB method
plt.plot(x_axis, y_axis)
```

```python
# Create a plot using the object-oriented method
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create with plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the line plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the line plot and add a label for the legend
plt.plot(x_axis, y_axis, marker="d", color="green", linewidth=2, label='Boston')

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0,45)

# Create a title
plt.title("PyBer Fare by Month")

# Add a grid
plt.grid()

# Add the legend.
plt.legend()
```

```python
# Create the bar plot
plt.bar(x_axis, y_axis)
```

```python
# Create the bar plot
plt.bar(x_axis, y_axis, color="green", label="Boston")

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()
```

```python
# Create the horizontal bar plot
plt.barh(x_axis, y_axis)

# Note that with this method, the x and y axes are swapped
```

```python
# Create the horizontal bar plot with swapped axes
plt.barh(y_axis, x_axis)
```

```python
# Create the plot with gca() ["get current axes"] function and invert_raxis() method
plt.barh(x_axis, y_axis, color="magenta", label="Boston")

# Add title
plt.title("Horizontal PyBer Chart")

# Add Axes Labels
plt.xlabel("Fare")
plt.ylabel("Month")

# Add a legend
plt.legend("Boston")

# Add axis inversion
plt.gca().invert_yaxis()
```

```python
# Create the plot using ax.plt() object oriented programming
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the horizontal plot using ax.plt() object oriented programming
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
# Create the horizontal plot with swapped axes using ax.plt() object oriented programming
fig, ax = plt.subplots()
ax.barh(y_axis, x_axis, color="cyan", label="Chicago")

# Add title
ax.set_title("PyBer Fares Per Month")

# Add labels
ax.set_xlabel("Months")
ax.set_ylabel("Fare($)")
```

```python
# Create the scatter plot
plt.plot(x_axis, y_axis, 'o')
```

```python
# Create the scatter plot using plt.graph()
plt.scatter(x_axis, y_axis, color="green", marker="d", s=100)
```

```python
# Create the bubble plot using plt.graph()
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Create a iteratable list to create larger size variations for the bubble plot
y_axis_larger = []
for value in y_axis:
    y_axis_larger.append(value*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Understand list comprehension
# new_list = [expression for item in list if conditional]
```

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Create the bubble plot using ax.graph() object-oriented programming
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis_larger, alpha=0.5, facecolor="red", edgecolor="cyan", label="Boston")

# Add a title
ax.set_title("Graph Title")

# Add labels
ax.set_xlabel("Months")
ax.set_ylabel("Fare($)")

# Add axes limits
ax.set_xlim("Jan", "Dec")
ax.set_ylim(0,50)

# Add legend
ax.legend("Boston")
```

```python
# Create a pie chart
# Notice how y_axis - your dependent variable - goes first and the pie pieces are labeled after your x axis - your independent variable
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Separate pie pieces by utilizing "explode" and explode_values

# Create a list for explode values
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

# Create the pie chart
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Adjust the pie chart figure size(s) - Must add this first btw !
plt.subplots(figsize=(8,8))

# Create a list for explode values
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

# Create the pie chart
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign colors to all twelve pie pieces

# Create a list for colors - same amount as x_axis components
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

# Create a list for explode values
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

# Adjust figure size
plt.subplots(figsize=(8,8))

# Create pie chart
plt.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, autopct="%1.1f%%")

# Show the plot
plt.show()
```

```python
# Create a pie chart using ax.graph() object oriented programming
fig, ax = plt.subplots()
ax.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Import statistics module / dependency
import statistics
```

```python
# Add error bars to line charts

# Create errorbar line chart
plt.errorbar(x_axis, y_axis, yerr=1.5)
```

```python
# Get the standard deviation of the values in the y_axis
stdev = statistics.stdev(y_axis)
stdev
```

```python
# Create an errorbar line chart with standard deviation as margin of error
plt.errorbar(x_axis, y_axis, yerr=stdev)
```

```python
# Create an errorbar line chart with stdev & capsize specification
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Create an errorbar line chart with atas using ax.graph() object oriented programming
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=5)
plt.show()
```

```python
# Adding error bars and stats to other charts

# Bar Chart
plt.bar(x_axis, y_axis, yerr=stdev, capsize=4)
plt.show()

# W/ ax.graph() object oriented programming
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
# Change major tick distance and setting using/by importing numpy

#import numpy
import numpy as np

# Create bar chart (horizontal)
plt.barh(x_axis, y_axis)

# Edit X-axis ticks - review this
plt.xticks(np.arange(0,51, step=5))

# Invert Axes using gca() & invert_yaxis()
plt.gca().invert_yaxis()
```

```python
# Change major tick distance and setting using/by importing numpy using ax.graph() object oriented programming
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, yerr=2, color="k", capsize=3)
ax.set_xticks(np.arange(0,51, step=5))

plt.show()
```

```python
# Change minor tick distance

from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
